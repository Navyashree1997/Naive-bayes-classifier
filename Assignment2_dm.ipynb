b] Build a vocabulary as list

```python
# def build_vocab(category,data):
vocab = {}
for i in train_data['New_Sentence']:
    ele=str(i).split()
    for j in ele:
        word=j
        if (word in vocab) :
            vocab[word]+=1
        else:
            vocab[word]=1
# print(vocab)
```

```python
for word in list(vocab):
    if vocab[word]<5:
        del vocab[word]
```
